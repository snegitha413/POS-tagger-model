```python
words = []
pos = []
X = []
Y = []
for i in range(len(df)):
  if df['UPOS'].iloc[i] != "_" and isNaN(df['UPOS'].iloc[i]) == False:
    words.append(df['WORD'].iloc[i])
    pos.append(df['UPOS'].iloc[i])
  elif isNaN(df['ID'].iloc[i]) == True:
    X.append(words)
    Y.append(pos)
    words = []
    pos = []

print(X)
```

```python
from keras.preprocessing.text import Tokenizer

word_tokenizer = Tokenizer()
word_tokenizer.fit_on_texts(X)
X_encoded = word_tokenizer.texts_to_sequences(X)

tag_tokenizer = Tokenizer()
tag_tokenizer.fit_on_texts(Y)
Y_encoded = tag_tokenizer.texts_to_sequences(Y)

print("** Raw data point **", "\n", "-"*100, "\n")
print('X: ', X[0], '\n')
print('Y: ', Y[0], '\n')
print()
print("** Encoded data point **", "\n", "-"*100, "\n")
print('X: ', X_encoded[0], '\n')
print('Y: ', Y_encoded[0], '\n')
```

```python
# to find max sequence length
def find_max_list(list):
    list_len = [len(i) for i in list]
    return (max(list_len))
```

```python
max_seq_length = find_max_list(X)
max_seq_length
```

```python
max_seq_length1 = find_max_list(X_validate)
max_seq_length1
```

```python
from tensorflow.keras.preprocessing.sequence import pad_sequences

X_padded = pad_sequences(X_encoded, maxlen=max_seq_length, padding='pre', truncating='post')
Y_padded = pad_sequences(Y_encoded, maxlen=max_seq_length, padding='pre', truncating='post')

print(X_padded[0], "\n"*3)
print(Y_padded[0])
```

```python
# word_tokenizer1 = Tokenizer()
# word_tokenizer1.fit_on_texts(X_validate)
X_validate_encoded = word_tokenizer.texts_to_sequences(X_validate)

# tag_tokenizer1 = Tokenizer()
# tag_tokenizer1.fit_on_texts(Y_validate)
Y_validate_encoded = tag_tokenizer.texts_to_sequences(Y_validate)

print("** Raw data point **", "\n", "-"*100, "\n")
print('X: ', X_validate[0], '\n')
print('Y: ', Y_validate[0], '\n')
print()
print("** Encoded data point **", "\n", "-"*100, "\n")
print('X: ', X_validate_encoded[0], '\n')
print('Y: ', Y_validate_encoded[0], '\n')
```

```python
X_validate_padded = pad_sequences(X_validate_encoded, maxlen=max_seq_length, padding='pre', truncating='post')
Y_validate_padded = pad_sequences(Y_validate_encoded, maxlen=max_seq_length, padding='pre', truncating='post')

print(X_validate_padded[0], "\n"*3)
print(Y_validate_padded[0])
```

```python
vocab_size = len(word_tokenizer.word_index) + 1
vocab_size
```

LSTM Model

```python
from sklearn.model_selection import train_test_split
from keras.utils.np_utils import to_categorical

Y = to_categorical(Y_padded)
Y_validate = to_categorical(Y_validate_padded)
X_train, X_validation, Y_train, Y_validation = train_test_split(X_padded, Y, test_size = 0.2)
X_test = X_validate_padded
Y_test = Y_validate
NUM_CLASSES = Y.shape[2]
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, LSTM, Embedding, SpatialDropout1D, Bidirectional, TimeDistributed, Conv1D, MaxPooling1D, SimpleRNN
```

```python
model = Sequential()
model.add(Embedding(input_dim = vocab_size, output_dim = 300, input_length  =  max_seq_length, trainable = True))
model.add(SimpleRNN(128, return_sequences=True, activation = "relu"))
model.add(LSTM(128, return_sequences=True, activation = "relu"))
model.add(TimeDistributed(Dense(NUM_CLASSES, activation='softmax')))
model.summary()
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['acc'])
```

```python
model_training = model.fit(X_train, Y_train, batch_size=128, epochs=25, validation_data=(X_validation, Y_validation))
```

```python
import matplotlib.pyplot as plt
accuracy = model_training.history['acc']
val_accuracy = model_training.history['val_acc']
epochs = range(1,26)
plt.plot(epochs, accuracy, 'g', label='Training accuracy')
plt.plot(epochs, val_accuracy, 'y', label='Validation accuracy')
plt.title('Training and Validation Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()
```

```python
loss_train = model_training.history['loss']
val_loss= model_training.history['val_loss']
epochs = range(1,26)
plt.plot(epochs, loss_train, 'g', label='Training loss')
plt.plot(epochs, val_loss, 'y', label='Validation loss')
plt.title('Training and Validation Loss')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
model.evaluate(X_test, Y_test)
```
